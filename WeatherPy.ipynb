```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of randum latitude and longitude combinations
lats = np.random.uniform(-90.000, 90.000, size=1500)
lngs = np.random.uniform(-90.000, 90.000, size=1500)
lat_lgns = zip(lats, lngs)
lat_lgns
```

```python
# Create a practice set of random latitude and longitude combinations.
x = [25.12903645, 25.92017388, 26.62509167, -59.98969384, 37.30571269]
y = [-67.59741259, 11.09532135, 74.84233102, -76.89176677, -61.13376282]
coordinates = zip(x, y)
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(coordinate[0], coordinate[1])
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lgns)
```

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
```

```python
# Use the tuple() function to display the latitude and longitude combination 
for coordinate in coordinates:
        print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name, citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create a list for holding the cities
cities = []

# Identify the nearest city for each lat/lng combo
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # If the city is unique, add to cities list
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

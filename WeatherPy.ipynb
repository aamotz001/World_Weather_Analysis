```python
# Create an empty list to hold the weather data.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# Loop through all the cities in the list
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
   
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count.
    record_count += 1
    
    # Run an API request for each of the cities
    try:
        
        # Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        
        # Convert the data to ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # Append the city info to city_data list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
    
    # If an error is experienced, skip the city
    except:
        print('City not found. Skipping... ')
        pass

# Indicate that data loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Import linear regression from the SciPy stats module
from scipy.stats import linregress
```

```python
# Create an equal number of latitudes and temperatures.
nlats = [42.5, 43.9, 8.1, 36.8, 79.9, 69.1, 25.7, 15.3, 12.7, 64.5]
temps = [80.5, 75.3, 90.9, 90.0, 40.4, 62.3, 85.4, 79.6, 72.5, 72.0]
```

```python
# Perform linear regression
(slope, intercept, r_value, p_value, std_err) = linregress(nlats, temps)

# Get the equation of the line
line_eq = 'y = ' + str(round(slope,2)) + 'x ' + str(round(intercept,2))
print(line_eq)
print(f'The p-value is {p_value:.3f}')
```

```python
# Calculate the regression line y values from the lope and interecept
regress_values = [(lat * slope + intercept) for lat in nlats]
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# create a scatter plot of the x and y vals
plt.scatter(nlats, temps)

# Plot the regression line with the x-values and the y coordinates based on intercept/slope
plt.plot(nlats,regress_values,"r")

# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Temp')
plt.show()
```

```python
# Perform linear regressiomn fpr max temp
(slope, intercept, r_value, p_value, std_err) = linregress(lats, max_temps)

# Calculate the regression line "y values" from the slope and intercept.
regress_values = lats * slope + intercept

# Get the equation of the line.
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Create a scatter plot of the x and y values.
plt.scatter(lats, max_temps)

# Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
plt.plot(lats, regress_values, "r")

# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Temp')
plt.show()
```

```python
# Create a fxn to make these linear regressions on weather data and plot
def plot_linear_regression(x_vals, y_vals, title, y_label, text_coords):
    
    # Do the linear regression
    (slope, intercept, r_value, p_value, std_err) = linregress(x_vals, y_vals)
    
    # Calculate the regression line "y values" from the slope and intercept.
    regress_values = x_vals * slope + intercept
    
    # Get the equation of the line.
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    # Create a scatter plot of the x and y values.
    plt.scatter(x_vals, y_vals)
    
    # Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
    plt.plot(x_vals, regress_values, "r")
    
    # Annotate the text for the line equation.
    plt.annotate(line_eq, text_coords, fontsize=15, color="red")
    plt.title(title)
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    plt.show()
```

```python
 plot_linear_regression(lats, max_temps, 'Linear Regression on Northern Hemisphere \nfor Max Temp', 'y', (-40, 10))
```

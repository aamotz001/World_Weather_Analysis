# Deliverable 1: Retrieve Weather Data

```python
# Import dependencies
import numpy as np
from citipy import citipy
from datetime import datetime
import requests
import pandas as pd

# Import the API key
from config import weather_api_key

# Verify API key was imported
weather_api_key
```

### Create a set of 2000 random latitudes and longitudes:

```python
# Generate 2000 latitude and longitude balues with the random.uniform fxn
lat_vals = np.random.uniform(-90.000, 90.000, size=2000)
lng_vals = np.random.uniform(-90.000, 90.000, size=2000)
```

```python
# Create 2000 latitude/longitude pairs from the values created in 1
lat_lng_pairs = zip(lat_vals, lng_vals)
```

```python
# Add the pairs to a list
coords = list(lat_lng_pairs)
```

```python
# Check that there are 2000 coordinate pairs... 
len(coords)
```

```python
# ... and that the syntax is correct for each pair
coords[1]
```

### Get a list of cities from the citypy module

```python
# Create a list for holding the cities
city_list = []
```

```python
# Find the city nearest to each coordinate pair
for coordinate in coords:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # If the city is unique, add to cities list
    if city not in city_list:
        city_list.append(city)
```

```python
# Print the city count to confirm sufficient count
len(city_list)
```

### Perform an API call with OpenWeatherMap

```python
# Set up the URL for OpenWeatherMap
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Create an empty list to hold the weather data.
city_weather_data = []
# Print the beginning of the logging.
print("Beginning the Weather Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_i = 1
set_i = 1

# Loop through all the cities in the city list
for i, city in enumerate(city_list):
    
    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_i += 1
        record_i = 1
        
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
   
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_i} of Set {set_i} | {city}")
    
    # Add 1 to the record count.
    record_i += 1
    
    # Run an API request for each of the cities
    try:
        
        # Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather['weather'][0]['description']
        
        # Convert the data to ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # Append the city info to city_data list
        city_weather_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_description,
                          "Country": city_country})
    
    # If an error is experienced, skip the city
    except:
        print('City not found. Skipping... ')
        pass

# Indicate that data loading is complete
print("-----------------------------")
print("Ding! All done.      ")
print("-----------------------------")
```

### Add the API data to a new DataFrame

```python
# Add data to DataFrame
city_weather_data_df = pd.DataFrame(city_weather_data)

# Change the column order
new_cols = ['City','Country','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','Current Description']
city_weather_data_df = city_weather_data_df[new_cols]

# Format the columns 
city_weather_data_df['Lat'] = city_weather_data_df['Lat'].map('{:,.2f}'.format)
city_weather_data_df['Lng'] = city_weather_data_df['Lng'].map('{:,.2f}'.format)
city_weather_data_df['Humidity'] = city_weather_data_df['Humidity'].map('{:,.0f}'.format)

# Print the DataFrame head to verify
city_weather_data_df.head(10)
```

### Export the DataFrame as a CSV file

```python
# Create the outpit file (CSV)
output_data_file = 'WeatherPy_Database.csv'

# Export the city data to a CSV
city_weather_data_df.to_csv(output_data_file, index_label = 'City_ID')
```
